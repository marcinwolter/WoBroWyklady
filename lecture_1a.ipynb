# Problem komiwojażera (Traveling Salesman Problem, TSP)



https://www2.seas.gwu.edu/~simhaweb/champalg/tsp/tsp.html

**Komiwojażer ma odwiedzić $n$ danych miast tak, aby wrócić do miasta startowego i aby całkowita droga była jak najkrótsza.**

(odległość/czas/koszt/...)

Matematycznie: znaleźć minimalny cykl Hamiltona w grafie pełnym z wagami (zob. rys. powyżej).

Jest to problem klasy NP, tj. niewykonalny w czasie wielomianowym, a wykonalny w czasie wykładniczym (zob. mój wykład z matematyki dyskretnej). Algorytm *dynamiczny* (Held–Karp, idea podobna do poznanej na Matematyce dyskretnej metody "dziel i rządź") rozwiązuje go w czasie $O(n^{2}2^{n})$.

https://www.geeksforgeeks.org/travelling-salesman-problem-set-1/

### Euklidesowa odległosć między dwoma miastami i oraz j

W ogólności, odległości mogą byc dowolnymi liczbami i nie muszą nawet dawać metryki (symetria, nierówność trójkąta). Mamy wówczas niesymetryczny lub niemetryczny problem komiwojażera.

$d_{ij}=\sqrt{(x_i-x_j)^2+(y_i-y_j)^2}$

```python
def dist(i,j):
    return np.sqrt((cities[i,0]-cities[j,0])**2 + (cities[i,1]-cities[j,1])**2)
```

```python
sequence[0], sequence[1]
```

```python
dist(sequence[0],sequence[1])
# por. z powyższym rysunkiem
```

```python
# długość drogi
def len_path(path):
    return sum([dist(path[i],path[i+1]) for i in range(n)])
```

```python
lenpa=np.round(len_path(sequence),3) # zaokrąglenie
lenpa
```

```python
plt.figure(figsize=(5,5))
plt.title("Losowa droga",fontsize=16) 
plt.xlim(-.1,1.1)
plt.ylim(-.1,1.1)

plt.scatter(cities[:,0],cities[:,1],c='red', s=20)

plt.plot(tour[:,0],tour[:,1],c='blue')

plt.annotate(lenpa, (.8,.95),fontsize=14)

plt.xlabel('$x$',fontsize=18)
plt.ylabel('$y$',fontsize=18);
```

## Algorytm najbliższego sąsiada (zachłanny)

**Algorytm zachłanny (greedy algorthm) optymalizuje efekt w najblizszym kroku, ale nie myśli o dalszej przyszłości!**

[Wiki] Algorytm zachłanny (ang. greedy algorithm) – algorytm, który w celu wyznaczenia rozwiązania w każdym kroku dokonuje zachłannego, tj. najlepiej rokującego w danym momencie wyboru rozwiązania częściowego. Innymi słowy algorytm zachłanny nie dokonuje oceny czy w kolejnych krokach jest sens wykonywać dane działanie, dokonuje decyzji lokalnie optymalnej, dokonuje on wyboru wydającego się w danej chwili najlepszym, kontynuując rozwiązanie podproblemu wynikającego z podjętej decyzji.

```python
# algorytm najbliższego sąsiada

def near_nei(ini): # ini - miasto startowe
    nn=[] # pusta lista dla sekwencji odwiedzonych miast (trasa)
    to_visit=np.array([i for i in range(n)]) # miasta do odwiedzenia
    ind=ini # zaczynamy od miasta startowego
    print('jest w ', ind)
    nn.append(ind) # dodajemy do trasy
    to_visit=np.delete(to_visit,ind) # usuwamy z listy miast do odwiedzenia
    print('do odwiedzenia: ', to_visit)
    for k in range(n-1): # pętla po liczbie miast do odwiedzenia
        ind0=np.argmin([dist(ind,i) for i in to_visit]) # położenie najbliższego miasta 
        ind=to_visit[ind0] # numer miasta do odwiedzenia
        print('jest w ',ind)
        nn.append(ind) # dodanie do trasy
        to_visit=np.delete(to_visit,ind0) # usunięcie z listty miast do odwiedzenia
        print('do odwiedzenia: ', to_visit)
    nn.append(ini) # dodanie do końca trasy miasta początkowego
    return nn     
```

```python
# to samo bez pisania
def near_nei(ini):
    nn=[]
    to_visit=np.array([i for i in range(n)])
    ind=ini
    nn.append(ind)
    to_visit=np.delete(to_visit,ind)
    for k in range(n-1):
        ind0=np.argmin([dist(ind,i) for i in to_visit])
        ind=to_visit[ind0]
        nn.append(ind)
        to_visit=np.delete(to_visit,ind0)
    nn.append(ini)
    return nn
```

## Dylemat mrówki

Na początek prosty cząstkowy problem. Załóżmy, że mrówka z danego miasta może pójść do 4 innych miast, a prawdopodobieństwa wyboru drogi do tych miast wynoszą odpowiednio [0.5, 0.2, 0.1, 0.2] (sumują się do 1). Na początek robimy tablicę skumulowanych prawdopodobieńst *(NB w teorii prawdopodobieństwa nazywa się to dystrybuantą).*

```python
# skumulowana suma tablicy
from itertools import accumulate
```

```python
tab_p=np.array([0.5, 0.2, 0.1, 0.2])
cum_p=np.array(list(accumulate(tab_p)))
np.round(cum_p,2)
```

```python
np.sum(np.heaviside(rd-cum_p,0)).astype(int) # konwersja do typu integer
```

```python
def ind_prob(tab_p):
    cum_p=np.array(list(accumulate(tab_p)))
    return np.sum(np.heaviside(random.random()-cum_p,0)).astype(int)
```

```python
ind_prob(tab_p)
```

```python
test_tab=np.array([ind_prob(tab_p) for _ in range(10000)])
```

## Wracamy do algorytmu mrówkowego dla TSP

```python
# parametry modelu dot. prawdopodobieństwa wybory drogi przez mrówkę
alpha = 1.15;
beta = 1;
h = 0.03;
popsize = 10;
sc = n/popsize;
```

```python
# zwraca (losowo) miasto do odwiedzenia z listy miast togo, jeśli mrówka jest w mieście i 
def ac_next(i, togo):
    p =np.array([fero[i, j]**alpha/dis_tab[i, j]**beta for j in togo]) # wagi
    su=np.sum(p) # suma wag
    p=p/su # prawdopodobieństwa (znormalizowane do 1)
    return togo[ind_prob(p)]
```
